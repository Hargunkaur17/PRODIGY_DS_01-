```python
import matplotlib.pyplot as plt

# Sample data representing the distribution of genders in a population
genders = ['Male', 'Female', 'Non-Binary']
counts = [500, 600, 200]

# Create a bar chart
plt.bar(genders, counts)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in a Population')

# Display the chart
plt.show()
```
